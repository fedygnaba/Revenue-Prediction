### Preparing the DATA


```python
features=df.drop('Revenue',axis=1)
labels=df['Revenue']
```

```python
labels
```

```python
# categorical data 
features = pd.get_dummies(features, columns=['VisitorType'])
```

```python
features.sample(7)
```

```python
# Method 2: Normalisation, (entre 0 et 1)
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
features[['id','ProductRelated','Administrative',]] = mms.fit_transform(features[['id','ProductRelated','Administrative']])
#labels = mms.fit_transform(dataset[['Revenue']])
```

```python
features
```

```python
#data training and testing 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, y_train)
```

```python
print(  round(log.score(X_train, y_train),4) )
print(  round(log.score(X_test, y_test)  ,4) )
# The right fit
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = log.predict(X_test)
```

```python
from sklearn.metrics import classification_report
# all values are balanced
print(classification_report(y_test,y_pred))
```
